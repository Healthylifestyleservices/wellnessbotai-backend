## Establish price anchoring logic

### Subtask:
Outline the strategy for presenting the tiered pricing to guide potential clients towards their desired level of investment and transformation.


**Reasoning**:
Describe the strategy for presenting the tiered pricing, including price anchoring, the order of presentation, highlighting value propositions, framing the highest tier, and positioning the lowest tier.



```python
# 1. Describe Price Anchoring and Application
price_anchoring_description = """
## Price Anchoring Strategy

Price anchoring is a psychological pricing strategy where a high-priced option is presented first or prominently to make subsequent, lower-priced options seem more reasonable or attractive by comparison. In the context of The Sacred Reclamation, the highest tier, **Cosmic Embodiment ($22,222)**, will serve as the primary anchor. By presenting this premium, all-inclusive option first, the **Solar Radiance ($11,111)** and **Lunar Bloom ($5,555)** tiers will appear significantly more accessible and lower in investment, even though they represent high-ticket prices in absolute terms. This strategy leverages the contrast effect, making the middle and lower tiers feel like a greater value relative to the anchor.
"""

# 2. Explain the Order of Presentation
presentation_order_explanation = """
## Order of Presentation

The tiers will be presented in descending order of investment:

1.  **Cosmic Embodiment ($22,222)**
2.  **Solar Radiance ($11,111)**
3.  **Lunar Bloom ($5,555)**

This order is crucial for the price anchoring strategy to be effective. Presenting the most expensive tier first establishes a high benchmark in the potential client's mind. Subsequent tiers are then evaluated against this anchor, making their prices seem less daunting and highlighting their relative affordability.
"""

# 3. Detail Highlighting Value Proposition
value_proposition_highlight = """
## Highlighting Value Propositions

To effectively support the price anchoring, the value proposition of each tier will be clearly articulated, emphasizing the increasing benefits and exclusivity as the investment level rises:

*   **Cosmic Embodiment:** Positioned as the ultimate, transformative experience offering complete grid restoration and full divine embodiment. The deliverables will highlight the maximum level of personalized support (6x 1:1 deep dive sessions), exclusive access (lifetime materials, private sisterhood), and unique energetic prescriptions and initiation. The language will emphasize deep, lasting, and unparalleled transformation.

*   **Solar Radiance:** Presented as a significant, accelerated transformation journey. While less investment than Cosmic Embodiment, the deliverables will showcase substantial personalized support (3x 1:1 sessions), customized protocols, and advanced practices that go significantly beyond the foundational tier. It will be framed as the ideal choice for those ready for profound shifts and personalized attention, making it feel like a robust, high-value option that is still more attainable than the top tier.

*   **Lunar Bloom:** Described as the sacred entry point into the transformation. The deliverables will focus on providing foundational knowledge, group support, and essential practices for initial clearing and alignment. It will be presented as a powerful beginning, making the higher tiers appear as the natural progression for deeper, more personalized, and accelerated results. The value will be highlighted by emphasizing access to the core program container and initial tools for transformation at the lowest investment level.
"""

# 4. Include Specific Language/Framing
framing_language = """
## Framing Language

*   **For Cosmic Embodiment:** Use language that evokes exclusivity, ultimate potential, divine alignment, and complete restoration. Phrases like "The pinnacle of transformation," "Claim your divine sovereignty," "Unlock your ultimate energetic blueprint," "An unparalleled journey into sacred embodiment," "The most comprehensive and personalized path to grid restoration."

*   **For Solar Radiance:** Frame it as the accelerated path, a significant leap forward with personalized support. Phrases like "Illuminate your path with personalized guidance," "Accelerate your energetic purification," "Step into amplified radiance," "Deepen your transformation with dedicated support," "The most popular choice for profound and personalized results."

*   **For Lunar Bloom:** Position it as the gentle beginning, the sacred entry point. Phrases like "Begin your sacred alignment," "Embrace the gentle unfolding," "Your foundational journey into energetic grace," "A beautiful entry point into The Sacred Reclamation," "Start your transformation with foundational support."
"""

# 5. Explain Lowest Tier's Role
lowest_tier_role = """
## Role of the Lowest Tier (Lunar Bloom)

The Lunar Bloom tier serves several key functions within the price anchoring strategy:

*   **Entry Point:** It provides an accessible entry point for potential clients who resonate with the program's essence but are not yet ready or able to commit to the higher investment levels. This allows them to experience the sacred container and foundational principles.
*   **Value Reinforcement:** By clearly outlining the deliverables in Lunar Bloom, it highlights what is *not* included compared to the higher tiers (e.g., personalized 1:1 sessions, advanced customizations, exclusive access). This contrast makes the added value and exclusivity of Solar Radiance and Cosmic Embodiment more apparent and desirable.
*   **Anchor Reinforcement:** While the highest tier is the primary anchor, the lowest tier also helps to frame the middle tier (Solar Radiance). Compared to Lunar Bloom, Solar Radiance offers significant upgrades (personalized sessions, advanced practices), making the jump in investment feel justified and the middle tier appear as a substantial step up in value.
"""

# Combine all sections
pricing_strategy_outline = (
    price_anchoring_description +
    presentation_order_explanation +
    value_proposition_highlight +
    framing_language +
    lowest_tier_role
)

# Display the outline
print(pricing_strategy_outline)
```
